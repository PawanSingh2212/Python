# Bar Plot

```python
import matplotlib.pyplot as plt

x= ["pyhton", "c","c++","Java"]
y=[85,70,60,82]

plt.bar(x,y)

plt.show()
```

####   Labeling , Fontsize , Width and colour

```python
x= ["pyhton", "c","c++","Java"]
y=[85,70,60,82]

plt.xlabel("Coding language", fontsize=10)
plt.ylabel("% used",fontsize=10)
plt.title("usage of coding languages",fontsize=15)
plt.bar(x,y,width=0.4,color=["c","y","b","r"])
plt.show()
```

- Align , Center

```python
x= ["pyhton", "c","c++","Java"]
y=[85,70,60,82]

plt.bar(x,y,align="edge")
plt.show()
```

```python
x= ["pyhton", "c","c++","Java"]
y=[85,70,60,82]

plt.bar(x,y,align="center")

plt.show()
```

####  Edge colour , Linewidht, Alpha

```python
x= ["pyhton", "c","c++","Java"]
y=[85,70,60,82]

plt.bar(x,y,edgecolor="r")

plt.show()
```

```python
# linewidht

x= ["pyhton", "c","c++","Java"]
y=[85,70,60,82]

plt.bar(x,y,edgecolor="r",linewidth=10)

plt.show()
```

```python
# alpha

x= ["pyhton", "c","c++","Java"]
y=[85,70,60,82]

plt.bar(x,y,color=["r","g","c","orange"],alpha=0.4)

plt.show()
```
