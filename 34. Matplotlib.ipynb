# Pie Plot 

```python
x=[10,20,30,40]
y=['c','c++','java','python']

plt.pie(x)
plt.show()
```

```python
# Label

x=[10,20,30,40]
y=['c','c++','java','python']

plt.pie(x,labels=y)
plt.show()
```

```python
# Explode

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.1,0.0]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
# colours
x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]
c=['c','b','y','pink']
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
# Autopct for percentage

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.1,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%")
plt.show()
```

```python
x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.1,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.3f%%")  # #.3 means numbers after decimal
plt.show()
```

```python
# Shadow

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.1,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True)
plt.show()
```

```python
# Radius 

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",radius=1.5)
plt.show()
```

```python
# Label distance

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.1,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",labeldistance=1.6)
plt.show()
```

```python
# starting angle 

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",startangle=90)
plt.show()
```

```python
# textprops / Fontsize 

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",textprops={"fontsize":15,})
plt.show()
```

```python
# Counterclock


x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",counterclock=False)
plt.show()
```

```python
# wedgeprops 

x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,wedgeprops={"linewidth":4,"width":1,"edgecolor":"m"})
plt.show()
```

```python
# Center
x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",center=(4,4))
plt.show()
```

```python
# Rotate label 


x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",rotatelabels=True)
plt.show()
```

```python
# Title 


x=[10,20,30,40]
y=['c','c++','java','python']

ex=[0.4,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",)

plt.title("PIE CHART")

plt.show()
```

```python
# Legend , location 

x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",)

plt.legend(loc=3) # 1,2,3,4 places

plt.title("PIE CHART")
plt.show()
```

```python
 # Dot Pie chart 

plt.pie([1])
plt.show()
```

# Stem plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y)
plt.show()
```

```python
# Linefmt

x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y,linefmt=":")
plt.show()
```

#### Linefmt 

- '_' for solid line
- '--' for Dashed line
- '-.' for  dash-dot line
- ':' for dootted line 

```python
# Markerfmt  color and shape 

x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y,markerfmt='r*')
plt.show()
```

```python
# Bottom line 

x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y,bottom=2)  # adjusting the middle line
plt.show()
```

```python
# Basefmt

x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y,basefmt='g')
plt.show()
```

```python
# Label 

x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y,basefmt='g',label='Python')
plt.legend()
plt.show()
```

```python
# 

x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y,use_line_collection=True)
plt.show()
```

```python
# Orientation

x=[1,2,3,4,5,6]
y=[-3,3,-1,2,3,4]

plt.stem(x,y,basefmt='g',orientation="horizontal")
plt.show()
```

# Box Plot & Whisker plot 

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x)
plt.show()
```

```python
# notch

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,notch=True)
plt.show()
```

```python
# Horizontal box plot

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,vert=False) # Vertical false
plt.show()
```

```python
# Width 

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,widths=0.8)
plt.show()
```

```python
# Label

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=['Python'])  # label data should be always list 
plt.show()
```

```python
# Patch_artist

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,patch_artist=True)
plt.show()
```

```python
# Showmeans

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True)
plt.show()
```

```python
# Whis

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,whis=2)  # whisk joined to outlier
plt.show()
```

```python
# Sym

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,sym='r+')  # Shape and color of outlier
plt.show()
```

```python
# Boxprops , capprops , whiskerprops

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,boxprops=dict(color='r'),capprops=dict(color='g'),whiskerprops=dict(color='c'))
plt.show()
```

```python
#flierprops

x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,flierprops=dict(markeredgecolor='y'))
plt.show()
```

```python
# Multiple box plot together 

x=[10,20,30,40,50,60,70,120]
y=[10,20,40,50,30,60,90,150]
plt.boxplot(x)
plt.boxplot(y)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
x1=[10,20,40,50,30,60,90,150]

y=[x,x1]

plt.boxplot(y,labels=['two', 'box'])
plt.show()
```

# Stack plot 

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[3,1,4,2,1]
area2=[5,6,7,3,2]
plt.stackplot(x,area,area1,area2,)
plt.show()
```

```python
x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[3,1,4,2,1]
area2=[5,6,7,3,2]
l=['area1','area2','area3']
plt.legend() # legend should always after plotting the graph
plt.stackplot(x,area,area1,area2,labels=l)
plt.show()
```

```python
# legends

x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[3,1,4,2,1]
area2=[5,6,7,3,2]
l=['area1','area2','area3']

plt.stackplot(x,area,area1,area2,labels=l)
plt.legend(loc=1)
plt.show()
```

```python
# colours

x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[3,1,4,2,1]
area2=[5,6,7,3,2]
l=['area1','area2','area3','area4']
c=['m','pink','y',]
plt.stackplot(x,area,area1,area2,labels=l,colors=c)
plt.legend()
plt.show()
```

```python
# baseline

x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[3,1,4,2,1]
area2=[5,6,7,3,2]
l=['area1','area2','area3']

plt.stackplot(x,area,area1,area2,labels=l,baseline='sym') # zero,wiggle, 
plt.legend()
plt.show()
```

```python
# gris, label , title 

x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[3,1,4,2,1]
area2=[5,6,7,3,2]
l=['area1','area2','area3']

plt.stackplot(x,area,area1,area2,labels=l)
plt.title("PYTHON")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## Fill between Plot 

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')

plt.fill_between(x,y)

plt.title('Python')
plt.xlabel('x-Axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')

plt.fill_between(x=[2,4],y1=2,y2=4,color='g')

plt.title('Python')
plt.xlabel('x-Axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Where argument 

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r',alpha=0.4)

plt.fill_between(x,y,color='g',where=(x>=2)&(x<=4),alpha=0.7)

plt.title('Python')
plt.xlabel('x-Axis')
plt.ylabel('y-axis')
plt.show()
```

## Saving Graphs
- savefig

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig('line')

plt.show()
```

```python
#dpi (for graph quality ), facecolor

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig('line',dpi=2000,facecolor='g')

plt.show()
```

```python
# saving graph on pdf format

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig('line.pdf')

plt.show()
```

```python
# Transparent 

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig('line2',dpi=2000,facecolor='b',transparent=True)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig('line3',dpi=2000,facecolor='b',transparent=True,bbox_inches="tight")

plt.show()
```

## Changes in axis 

### Xticks , yticks ,xlim , ylim, axis

```python
import matplotlib.pyplot as plt 
```
